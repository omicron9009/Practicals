```python
board=[ ['n','n','n'] ,
        ['n','n','n'] , 
        ['n','n','n'] ]
def print_board(board) :
    for i in board : 
        print(i)

def check(board):
    for i in range(len(board)): 
        if((board[i][0]=='x' and board[i][1]=='x'and  board[i][2]=='x') or (board[i][0]=='o' and board[i][1]=='o' and board[i][2]=='o') ):
            print("win","row")
            if((board[i][0]=='x' and board[i][1]=='x'and  board[i][2]=='x')):
                print("x win")
            else :
                print("o win")
            return 'w'
        
    for i in range(len(board)): 
        if((board[0][i]=='x' and board[1][i]=='x'and  board[2][i]=='x') or (board[0][i]=='o' and board[1][i]=='o'and  board[2][i]=='o') ):
            print("win","col")
            if((board[0][i]=='x' and board[1][i]=='x'and  board[2][i]=='x')):
                print("x win")
            else :
                print("o win")
            return 'w'
        
    if((board[0][0]=='x' and board[1][1]=='x'and  board[2][2]=='x') or (board[0][0]=='o' and board[1][1]=='o'and  board[2][2]=='o') ):
        print("win","dia 1")
        if((board[0][0]=='x' and board[1][1]=='x'and  board[2][2]=='x')):
            print("x win")
        else :
            print("o win")
        return 'w'
    
    if((board[0][2]=='x' and board[1][1]=='x'and  board[2][0]=='x') or (board[0][2]=='o' and board[1][1]=='o'and  board[2][0]=='o') ):
        print("win","dia 2")

        if((board[0][2]=='x' and board[1][1]=='x'and  board[2][0]=='x')):
            print("x win")
        else :
            print("o win")
        return 'w'
    for i  in board:
        if 'n' in i :
            return 'c'
    
    return "d"

def play():
    while True : 

        print("enter x (r,c)")
        x,y = map(int,input().split())
        print(x,y)
        if(x < 3 and y < 3) :
            if(board[x][y]=='n'):
                board[x][y]='x'
            else : 
                print("already marked place")
                continue
        else :
            print("invalid input ")
            continue

        print_board(board)

        print("enter o (r,c)")
        x,y = map(int,input().split())
        print(x,y)
        if(x < 3 and y < 3) :
            if(board[x][y]=='n'):
                board[x][y]='o'
            else : 
                print("already marked place")
                continue
        else :
            print("invalid input ")
            continue

        print_board(board)

        res=check(board)
        if(res=='w'):
            return
        if(res=='c'):
            continue
        if(res=='d'):
            print("Its a draw!")
            return 
        
        
play()
```
